```python
import pandas as pd
```

```python
path = "landmarks_dataset_M_A_H_W.csv"
df = pd.read_csv(path) 
print(df)
columns_with_negatives = df.columns[(df < 0).any()] 
print(columns_with_negatives)
print(len(columns_with_negatives))
```

```python
columns_to_drop = list(columns_with_negatives)
nonminus= []
for idx in columns_to_drop:
    col = int(idx)
    if col % 2==0:
        nonminus.append(str(col + 1))
    else:
        nonminus.append(str(col + 1))
```

```python
realCol= list(columns_to_drop) + nonminus
print(realCol)
print(len(realCol))
df = df.drop(columns = realCol)
```
